```python
from rank_bm25 import BM25Okapi


# Function to fetch documents from the vector store
def fetch_candidates_from_vectorstore(vector_store, query, top_n=10):
    results = vector_store.similarity_search(query, k=top_n)
    return results  # Returns a list of Document objects

# BM25-based re-ranking function
def bm25_rerank(query, retrieved_docs, top_n=5):
    # Extract text content from retrieved documents
    tokenized_docs = [doc.page_content.split(" ") for doc in retrieved_docs]
    tokenized_query = query.split(" ")

    # Initialize BM25 model
    bm25 = BM25Okapi(tokenized_docs)

    # Score documents based on the query
    scores = bm25.get_scores(tokenized_query)

    # Pair documents with scores and sort them
    scored_docs = sorted(
        zip(retrieved_docs, scores),
        key=lambda x: x[1],
        reverse=True
    )

    # Return top N re-ranked documents
    return [doc for doc, score in scored_docs[:top_n]]
```

```python
query = 'Tinh thần tiên phong, gương mẫu cộng bao nhiêu điểm'
# Retrieve initial candidates
retrieved_docs = fetch_candidates_from_vectorstore(vector_store, query, top_n=10)

# Perform BM25 re-ranking
re_ranked_docs = bm25_rerank(query, retrieved_docs, top_n=3)
```

```python
# Print results
print("Top Re-Ranked Documents (BM25):")
for doc in re_ranked_docs:
    print("-", doc.page_content)
```
