```python
import sqlite3
import pandas as pd
import os
```

```python
def read_sql_query(sql, db):
    conn = sqlite3.connect(db)
    cur = conn.cursor()
    cur.execute(sql)
    rows = cur.fetchall()
    for row in rows:
        print(row)
    conn.close()
    
```

```python
path_db = 'university_admissions.db'
df = read_sql_query('SELECT * FROM admission_methods;', path_db)
```

```python
query = f"SELECT * \
FROM admission_scores \
WHERE admission_method_id = 1 \
     AND year = 2021 \
     AND university_code = 'TDTU' \
     AND major_name LIKE '%Khoa học máy tính%';"

# Connect to the database and execute the query
with sqlite3.connect("university_admissions.db") as conn:
    df = pd.read_sql_query(query, conn)

# Display the filtered result
df
```
