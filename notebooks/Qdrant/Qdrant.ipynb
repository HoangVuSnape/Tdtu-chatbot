```python
# Set up collection name and vector parameters
os.environ['QDRANT_COLLECTION'] = "test_4"

# Define the collection configuration (size should match the model output, e.g. 768 for the model used here)
collection_config = qdrant_client.http.models.VectorParams(
    size=768,  # Size should match the output of your embedding model (768 in this case)
    distance=qdrant_client.http.models.Distance.COSINE
)

# Initialize Qdrant client
client = qdrant_client.QdrantClient(
    os.getenv("QDRANT_HOST"),
    api_key=os.getenv("QDRANT_API_KEY")
)

# Function to check if a collection exists and create it if necessary
def create_or_update_collection():
    collection_exists = client.collection_exists(collection_name=os.getenv("QDRANT_COLLECTION"))
    if collection_exists:
        print(f"Collection {os.getenv('QDRANT_COLLECTION')} already exists.")
    else:
        # Create the collection if it doesn't exist
        client.create_collection(
            collection_name=os.getenv("QDRANT_COLLECTION"),
            vectors_config=collection_config
        )
        print(f"Collection {os.getenv('QDRANT_COLLECTION')} created successfully.")

# Initialize the text splitter to handle large documents (split into chunks)
text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=100)

# Function to process files and upsert to Qdrant
def process_and_upsert_to_qdrant(folder_path):
    # Get all .txt files in the folder
    txt_files = glob.glob(os.path.join(folder_path, "*.txt"))
    print(f"Found {len(txt_files)} .txt files in the folder.")

    for file_path in txt_files:
        print(f"Processing file: {file_path}")

        # Read the content of the file
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()

        # Split the text into manageable chunks
        documents = [Document(page_content=text)]
        chunks = text_splitter.split_documents(documents)

        # Extract chunk texts
        chunk_texts = [chunk.page_content for chunk in chunks]
        print(f"Number of chunks created: {len(chunk_texts)}")

        # Generate embeddings for the chunks
        embeddings = embedding_model.encode(chunk_texts)
        print(f"Generated embeddings of shape: {embeddings.shape}")

        # Prepare points for upsert into Qdrant
        points = [
            {
                'id': str(uuid.uuid4()),   # Create a unique ID for each chunk (combining file name and chunk index)
                'vector': embedding,  # The embedding vector (size 768)
                'payload': {'text': chunk_texts[i], 'filename': os.path.basename(file_path)}  # Optional: store the text and filename as metadata
            }
            for i, embedding in enumerate(embeddings)
        ]

        # Upsert the points into Qdrant
        client.upsert(collection_name=os.getenv("QDRANT_COLLECTION"), points=points)
        print(f"Successfully upserted {len(points)} points from file {file_path}.")

# Run the function to process files and push to Qdrant
create_or_update_collection()  # Ensure collection exists or create it
folder_path = "/content/drive/MyDrive/TDTU_push /Data_QĐQC_TDTU/data_small_4_pdf"  # Specify your folder path
process_and_upsert_to_qdrant(folder_path)
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient
from qdrant_client.models import VectorParams, Distance

def search_query_in_qdrant(df, query_column, embedding_model, client, collection_name, limit=5):
    # Bước 1: Mã hóa tất cả các câu trong dataframe thành embedding
    df['embedding'] = df[query_column].apply(lambda x: embedding_model.encode([x])[0])

    # Bước 2: Tạo bảng kết quả để lưu các kết quả tìm kiếm
    search_results = []
    stt_counter = 0  # Đếm số thứ tự toàn cục
    # Bước 3: Thực hiện tìm kiếm trong Qdrant cho mỗi câu truy vấn
    for idx, row in df.iterrows():
        query_embedding = row['embedding']

        # Tìm kiếm trong Qdrant
        results = client.search(
            collection_name=collection_name,  # Tên collection cần tìm kiếm
            query_vector=query_embedding,  # Vector của truy vấn
            limit=limit,  # Số lượng kết quả cần lấy
            with_payload=True  # Lấy metadata (payload) kèm theo kết quả
        )

        # Bước 4: Lưu kết quả vào danh sách
        for i, result in enumerate(results):
            search_results.append({
                'sst': stt_counter,  # STT cho mỗi câu truy vấn, mỗi query sẽ có limit số thứ tự liên tiếp
                'query': row[query_column],  # Câu truy vấn
                'id': result.id,  # ID của kết quả tìm kiếm
                'filename': result.payload.get('filename', 'N/A'),  # Thông tin filename (nếu có)
                'result': result.payload.get('text', 'N/A'),  # Nội dung của kết quả
                'score': result.score  # Điểm số (cosine similarity)
            })
        stt_counter += 1

    # Bước 5: Tạo DataFrame từ kết quả tìm kiếm
    results_df = pd.DataFrame(search_results)

    return results_df
# Tìm kiếm và lấy kết quả từ Qdrant
result_table = search_query_in_qdrant(df, 'query', embedding_model, client, 'test_4', limit=5)

# In ra kết quả
result_table
```
